```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1,2,figsize=(10,5))
# Load the data from the .npy file
data = np.load('spec_test.npy')
x = np.linspace(0.0, 13677.9454908, 50)
y = np.linspace(24.5,25.31,80)
X,Y = np.meshgrid(x,y)
# Plot the 2D data
im = ax[0].pcolormesh(X,Y,data, cmap='viridis')
plt.colorbar(im,ax=ax[0])
ax[0].set_title('Plot of spec_test.npy')

im = ax[1].pcolormesh(X,Y,np.sqrt(data), cmap='viridis')
plt.colorbar(im,ax=ax[1])
ax[1].set_title('Plot of spec_test.npy')

plt.show()
```
